```python
class computing():

    def __init__(self, a, b):
        self.a = a
        self.b = b

    def addition(self):
        self.c = self.a + self.b
        return self.c

    def soustraction(self):
        self.d = self.a - self.b
        return self.d

    def multiplication(self, beta):
        c = computing(self.a, self.b).addition()
        d = computing(self.a, self.b).soustraction()
        e = (c*d)*(1-beta) # (self.c * self.d)*(1-beta)
        return e

    class c_or_d(computing):

        def __init__(self, vide):
            self.vide = vide

        def GGG(self):
            super(self.c)
            return self.c

        def HHH(self):
            super(self.d)
            return self.d
```
